```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("aapl_data.csv")
print(df.head())
```

```python
df=df.sort_values("date")
print(df.head())
```

```python
Y=df["target"]
X = df.drop(columns=["target", "date"])

print(X.head())
print(Y.head())
```

```python
K_fold=TimeSeriesSplit(n_splits=5,gap=10)
print(K_fold)

for i,(train_index,test_index) in enumerate(K_fold.split(df)):
    cut=int(len(train_index)*0.8)
    X_train = X.iloc[train_index[0] : train_index[0] + cut]
    X_val   = X.iloc[train_index[0] + cut : train_index[-1] + 1]
    X_test  = X.iloc[test_index[0] : test_index[-1] + 1]

    Y_train = Y.iloc[train_index[0] : train_index[0] + cut]
    Y_val   = Y.iloc[train_index[0] + cut : train_index[-1] + 1]
    Y_test  = Y.iloc[test_index[0] : test_index[-1] + 1]

    scaler = StandardScaler()

    scaler.fit(X_train)

    X_train_scaled = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
    X_val_scaled   = pd.DataFrame(scaler.transform(X_val),   columns=X_train.columns)
    X_test_scaled  = pd.DataFrame(scaler.transform(X_test),  columns=X_train.columns)

    print(X_train_scaled.head())
    
```
