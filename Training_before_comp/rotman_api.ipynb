```python
import time
import pandas as pd
from ritc import RIT

rit = RIT("IRJW2AFJ")

records = []
last_tick_recorded = -1
case_started = False

print("Waiting for case to become ACTIVE...")

while True:
    case = rit.get_case()

    # Wait for case to start
    if case.status != "ACTIVE":
        if case_started:
            # Case WAS running but now it's over — we're done
            print(f"Case ended at tick {last_tick_recorded}. Stopping collection.")
            break
        time.sleep(0.5)
        continue

    # Case is active
    case_started = True

    # Skip duplicate ticks
    if case.tick == last_tick_recorded:
        time.sleep(0.2)
        continue

    last_tick_recorded = case.tick

    # Snapshot
    for s in rit.get_securities():
        records.append({
            "tick":       case.tick,
            "ticker":     s.ticker,
            "last":       s.last,
            "bid":        s.bid,
            "ask":        s.ask,
            "bid_size":   s.bid_size,
            "ask_size":   s.ask_size,
            "volume":     s.volume,
            "position":   s.position,
            "unrealized": s.unrealized,
        })

    print(f"Tick {case.tick}/{case.ticks_per_period} captured")

    # Also stop if we hit the last tick explicitly
    if case.tick >= case.ticks_per_period:
        print("Final tick reached. Stopping collection.")
        break

# Save
df = pd.DataFrame(records)
df.to_csv("rit_data.csv", index=False)
print(f"\nSaved {len(df)} rows to rit_data.csv")
print(df.head())
```
